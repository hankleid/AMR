```python
from nptdms import TdmsFile
import numpy as np
from matplotlib import pyplot as plt
```

```python
files = ["20210318/20210318_800G_0.1mA_9deg_#1.tdms",
        "20210318/20210318_800G_0.1mA_9deg_#2.tdms",
        "20210318/20210318_800G_0.1mA_9deg_#3.tdms",
        "20210318/20210318_800G_0.1mA_9deg_#4.tdms"]

for i in range(len(files)):
    file = TdmsFile.read(files[i])
    data = file['Data']
    x = data['X']
    y = data['Y']
    plt.plot(x, y, label=i)
    
plt.title('AMR [800G, 0.1mA, 9deg step]', fontsize=16)
plt.xlabel('Angle (Degrees)', fontsize=16)
plt.ylabel('R (Ohms)', fontsize=16)
plt.xlim(0, 360)
plt.ylim(150, 250)
#plt.ylim(320, 420)
plt.legend()
#plt.savefig("20210318/20210318_Plot", dpi=200, facecolor='w', edgecolor='w', orientation='portrait')
```
